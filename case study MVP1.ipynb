```python
#Are the ratings affect the sales? 
project.groupby(['User_Score','Critic_Score'])[['Global_Sales']].sum().sort_values('Global_Sales',ascending=False)
```

```python
#What is the most widely Genre on each platform?
project.groupby(['Platform'])[['Genre']].max()
```

```python
#What is the Top five bestseller gems?
project.groupby(['Name'])[['Global_Sales']].sum().sort_values('Global_Sales',ascending=False).iloc[0:5].plot(kind='bar');
plt.title('Top five bestseller gems',fontsize = 20,weight='bold',color='black');
plt.xlabel('Name',fontsize = 10,weight='bold',color='black');
plt.ylabel('Global Sales',fontsize = 10,weight='bold',color='black');
```

```python
#Who is the top selling publisher?
project.groupby(['Publisher'])[['Global_Sales']].sum().sort_values('Global_Sales',ascending=False).iloc[0:3]
```

```python
project.groupby(['Publisher'])[['Global_Sales']].sum().sort_values('Global_Sales',ascending=False).iloc[0:3].plot(kind='bar');
plt.title('Top selling publisher',fontsize = 20,weight='bold',color='black');
plt.xlabel('Publisher',fontsize = 10,weight='bold',color='black');
plt.ylabel('Global Sales',fontsize = 10,weight='bold',color='black');
```
